# Task 1

```python
r1 = open("reviews1.txt", "r").read()
print(r1)
```

```python
r2 = open("reviews2.txt", "r").read()
print(r2)
```

```python
r3 = open("reviews3.txt", "r",encoding='utf-16').read()
print(r3)
```

```python
all_reviews = []
for i in [r1, r2, r3]:
    all_reviews.append(i.splitlines()[0])
    all_reviews.append(i.splitlines()[1])
    
len(all_reviews)
```

# Task2

```python
import re
```

```python
for r in range(len(all_reviews)):
    all_reviews[r] = re.sub(r'[^0-9\w\s\']', '', all_reviews[r].lower()).split()

all_reviews

```

```python
all_words = []
for r in all_reviews:
    for w in r:
        all_words.append(w)

print(all_words)
```

```python
from collections import Counter
```

```python
c = Counter(all_words)
print(c.most_common(5))
```

# Task3

```python
from nltk.stem import PorterStemmer
```

```python
stemmer = PorterStemmer()

for w in range(len(all_words)):
    all_words[w] = stemmer.stem(all_words[w])
```

```python
print(all_words)
```
