### Pre-processing

```python
my_punctuation = '!"$%&#()*+,-./:;<=>?[\\]^_`{|}~‚Ä¢'
    
def preprocess(text_string):
    space_pattern = '\s+'
    new_line = '\n+'
    giant_url_regex = ('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|'
        '[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+')
    mention_regex = '@[\w\-]+'
    non_word_char = '[^\w]'
    underscore = '_[\w]+'
    
    parsed_text = re.sub(space_pattern, ' ', text_string)
    parsed_text = re.sub(new_line, ' ', parsed_text)
    parsed_text = re.sub(giant_url_regex, '', parsed_text)
    parsed_text = re.sub(mention_regex, '', parsed_text)
    parsed_text = re.sub(non_word_char, ' ', parsed_text)
    parsed_text = re.sub(r"\bŸà(.*?)\b", r'\1', parsed_text)
    parsed_text = re.sub('([0-9]+)', '', parsed_text)
    parsed_text = re.sub(underscore, ' ', parsed_text)
    
    return parsed_text
```

```python
def deEmojify(text):
    regrex_pattern = re.compile(pattern = "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags = re.UNICODE)
    return regrex_pattern.sub(r'',text)
```

```python
trump['tidy_tweet'] = np.vectorize(preprocess)(trump['tweet'])
biden['tidy_tweet'] = np.vectorize(preprocess)(biden['tweet'])

trump['tidy_tweet'] = np.vectorize(deEmojify)(trump['tidy_tweet'])
biden['tidy_tweet'] = np.vectorize(deEmojify)(biden['tidy_tweet'])
```

```python
afin = {}

with open('AFINN.txt') as f: 
    sents = f.read()

for _, sentiment in enumerate(list(map(lambda x: x.split('\t'), sents.split('\n')[:-1]))):
    afin[sentiment[0]] = sentiment[1]
```

```python
words = list(afin.keys())
```

```python
def lexicon(a):
    s = sum(list(map(lambda x: int(afin[x])  , list(filter(lambda x: x in words, a)) )))
    
    return s
```

```python
trump['tweet_lexicon'] = trump['tidy_tweet'].apply(lambda x: lexicon(x))
```

```python
biden['tweet_lexicon'] = biden['tidy_tweet'].apply(lambda x: lexicon(x))
```
